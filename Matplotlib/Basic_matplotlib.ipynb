```python
import matplotlib.pyplot as plt
import numpy as np
```

# 선 그래프

1. plot() 함수
 - plt.plot(xArr, yArr, color, maker, linestyle)
    * color : r, g, b, c, y, k(black)...
    * maker: o, ^v<>, s, *...
    * linestyle : -, --, :, -.
  - plt.plot(xArr, yArr, format_string)
    *  색상 + 마커모양 + 라인스타일
    *  ro--  : 빨간색, 원형 마커, 실선
    *  b^-- : 파란색, 세모 마커, 실선





```python
years = np.array([50, 60, 70, 80, 90, 100, 110]) + 1900
gdp = np.array([67.0, 80.0, 257.0, 1686.0,6505, 11865.3, 22105.3])

# 선 그래프, x축 years, y축 gpd값
plt.plot(years, gdp, color = "g", marker = 'v', linestyle = ":")

# 제목 설정
plt.title("GDP per capita")

# y축 레이블 설정
plt.ylabel("dollars")
plt.show()

# png 이미지로 저장
# plt.savefig("제목.png", dpi = 600)
```

```python
# format_string

x = [x for x in range(-20, 20)]
y1 = [2 * t for t in x]
y2 = [t ** 2 + 5 for t in x]
y3 = [-t ** 2 - 5 for t in x]
y4 = [10 for _ in x]
plt.plot(x, y1, "r--", x, y2, "g^-", x, y3, "b*:", x, y4, "k--")
plt.axis([-30, 30, -30, 30])
plt.show()
```

범례 (legend)
 * plt.plot(x, y, label = "value")
 * plt.legend()

```python
import math
def fact(n):
  result = n
  n -= 1
  while n > 0:
    result *= n
    n -= 1
  return result
x = [x for x in range(20)]
y = [x ** 2 for x in range(20)]
z = [x ** 3 for x in range(20)]
w = [fact(x) for x in range(20)]
l = [math.log2(x) for x in range(1,21)]

plt.plot(x, z, label="O(N^3)")
plt.plot(x, x, label="O(N)")
plt.plot(x, w, label= "O(N!)")
plt.plot(x, y, label = "O(N^2)")
plt.plot(x, l, label = "O(log2N)")


plt.xlabel("x label")
plt.ylabel("y label")
plt.axis([0, 25, 0, 25])
plt.legend()
plt.show()
```

```python
import math

x = [i for i in range(360)]
y = [math.sin(math.radians(angle)) for angle in range(360)]
z = [math.cos(math.radians(angle)) for angle in range(360)]

plt.plot(x, y, 'b-', x, z, 'r-')
plt.title("SINE WAVE")
plt.show()
```

# 2. 막대 그래프

* bar() 함수
  * plt.bar(x, y)
* xticks() 함수
  * plt.xticts(x, ticks)

```python
plt.bar(range(len(years)), gdp)

plt.title("GDP per capita income")
plt.ylabel("dollars")

plt.xticks(range(len(years)), years)
plt.show()
```

다중 막대 그래프

```python
years = np.arange(65, 116, 10) + 1900
ko = np.random.randint(100, 30000,size=6)
jp = np.random.randint(800, 50000,size=6)
ch = np.random.randint(100, 8000,size=6)

x = np.arange(years.size)
plt.bar(x + 0.0, ko, width = 0.25, label = "KOREA")
plt.bar(x + 0.3, jp, width = 0.25, label = "JAPAN")
plt.bar(x + 0.6, ch, width = 0.25, label = "CHINA")
plt.legend()
plt.xticks(range(years.size), years)
plt.show()
```

# 3. 산포도(산점도) 그래프

```python
x = np.arange(20, 50)

y = x + 4 * np.random.randn(30)  # 표준 편차 4

plt.scatter(x, y)
plt.title("Real Age vs Physical Age")
plt.xlabel("Real Age")
plt.ylabel("Physical Age")
plt.show()
```

# 4. 파이 차트

pie() 함수


```python
ratio = np.random.rand(4) * 50
labels = ["Apple", "Banana", "Melon", "Grapes"]

plt.pie(ratio, labels=labels, autopct="%.3f%%")
plt.show()
```

```python
# 부채꼴 분리

ratio = [34,32,16,18]
labels = ["Apple", "Banana", "Melon", "Grapes"]
explode = [0, 0.1, 0, 0.5]
color = ["silver", 'gold', "whitesmoke", "lightgray"]
# 180도지난 후 반시계 방향으로 그림
plt.pie(ratio, labels = labels, autopct="%.1f%%", startangle=180, counterclock=True, explode = explode,
        shadow = True, colors=color)
plt.show()
```

# 5. 히스토그램

```python
weight = np.random.randint(50, 90, size = 22)
weight.sort()
print(weight)
plt.hist(weight, bins = 5) # 구간 개수

plt.show()
```

```python
# 다중 히스토그램
plt.hist(weight, label = "bins = 10", histtype = 'stepfilled')
plt.hist(weight, bins=30, label='bins = 30', histtype = "step")
plt.legend()
plt.show()
```

```python
n_bins = 10
x = np.random.randn(1000)
y = np.random.randn(1000)
plt.hist(x, n_bins, histtype = "bar", color = "red")
plt.hist(y, n_bins, histtype = "bar", color = "blue", alpha= 0.3)
plt.show()
```

```python
# 정규 분포 난수 배열 생성 함수

m1, s1 = 10, 2
m2, s2 = -6, 3

standard_Gauss = np.random.randn(10000)
Gauss1 = m1 + s1 * np.random.randn(10000)
Gauss2 = np.random.normal(m2, s2, 10000)

plt.figure(figsize = (10, 6))
plt.hist(standard_Gauss, bins = 50, alpha=0.4, label = f"mean : 0, Devi : 1")
plt.hist(Gauss1, bins = 50, alpha=0.4, label = f"mean : {m1}, Devi : {s1}")
plt.hist(Gauss2, bins = 50, alpha=0.4, label = f"mean : {m2}, Devi : {s2}")
plt.legend()
plt.show()
```

```python
# 균일 분포, 정규 분포, 표준 정규 분포
a = 2 * np.random.randn(10000) + 1
b = np.random.standard_normal(10000)
c = 20 * np.random.rand(10000) - 10

plt.hist(a, bins = 100, density=True, alpha=0.7, histtype = "step")
plt.hist(b, bins = 50, density=True, alpha=0.5, histtype = "stepfilled")
plt.hist(c, bins = 100, density=True, alpha=0.9, histtype = "step")

plt.show()
```
